````python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal
%

img=np.load("./u6c060507/digit9/digit9.npy")
edge=[[0,1,0],
      [1,-4,1],
      [0,1,0]]

plt.figure()
plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.axis('off')
plt.title('o i')

plt.subplot(1,2,2)
c_digit=signal.convolve2d(img,edge,boundary='symm',mode='same')
plt.imshow(c_digit,cmap='gray')
plt.axis('off')
plt.title('e i')
plt.show()
````
